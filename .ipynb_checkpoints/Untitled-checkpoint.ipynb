```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import row_number
from pyspark.sql.window import Window

spark = SparkSession.builder.appName("advsql").getOrCreate()

data = [
    (1, "Alice", "Engineering", 100000, "2023-01-15"),
    (2, "Bob", "Engineering", 95000, "2022-06-10"),
    (3, "Cathy", "HR", 80000, "2023-03-20"),
    (4, "David", "Engineering", 110000, "2021-09-05"),
    (5, "Eve", "HR", 85000, "2022-12-01")
]

df = spark.createDataFrame(data, ["id","name","dept","salary","hire_date"])
window_spec = Window.partitionBy("dept").orderBy("salary")
df_r = df.withColumn("row_id", row_number().over(window_spec))
df_r.show()
```
